```python
train_data['VAR2'].fillna(train_data['VAR2'].median(),inplace=True)
train_data['VAR3'].fillna(train_data['VAR3'].median(),inplace=True)
train_data['VAR4'].fillna(train_data['VAR4'].median(),inplace=True)
train_data['VAR5'].fillna(train_data['VAR5'].median(),inplace=True)
train_data['VAR6'].fillna(train_data['VAR6'].median(),inplace=True)
train_data['VAR7'].fillna(train_data['VAR7'].median(),inplace=True)
train_data['VAR8'].fillna(train_data['VAR8'].median(),inplace=True)
train_data['VAR9'].fillna(train_data['VAR9'].median(),inplace=True)
train_data['VAR10'].fillna(train_data['VAR10'].median(),inplace=True)
train_data['VAR11'].fillna(train_data['VAR11'].median(),inplace=True)
train_data['VAR12'].fillna(train_data['VAR12'].median(),inplace=True)
train_data['VAR13'].fillna(train_data['VAR13'].median(),inplace=True)
train_data['VAR15'].fillna(train_data['VAR15'].median(),inplace=True)
train_data['VAR16'].fillna(train_data['VAR16'].median(),inplace=True) 
train_data['VAR14']= train_data['VAR14'].astype('str') 
train_data['VAR14']=train_data['VAR14'].replace('.','1')
train_data['VAR14']= train_data['VAR14'].astype('int64') 
```

```python
train_data.head()
```

```python
nvars=['VAR2','VAR3','VAR4','VAR5','VAR6','VAR7','VAR8','VAR9','VAR10','VAR11','VAR12','VAR13','VAR14','VAR15','VAR16','VAR18','VAR19','VAR20']
X=train_data[nvars]
Y=train_data['VAR21']
```

```python
from sklearn.preprocessing import LabelBinarizer
encoder = LabelBinarizer()
transfomed_label = encoder.fit_transform(Y)
print(transfomed_label)
```

```python
xtrain, xvalid, ytrain, yvalid = train_test_split(X,Y, random_state=42, test_size=0.2)
```

```python
train_data['VAR21'].value_counts()
```

```python
from sklearn.ensemble import GradientBoostingClassifier#best results so far from Gradient boost
model = GradientBoostingClassifier(learning_rate=0.1,max_depth=5)
model.fit(xtrain, ytrain)
pred = model.predict(xvalid)
from sklearn.metrics import confusion_matrix, classification_report
print(confusion_matrix(yvalid, pred))
print(classification_report(yvalid, pred))
```

```python
#model = LogisticRegression(solver = 'lbfgs')
#model.fit(xtrain, ytrain)
```

```python
prediction = model.predict_proba(xvalid)
prediction_int = prediction.astype(np.int)
encoder.inverse_transform(prediction)
model.score(xvalid, yvalid)
```
